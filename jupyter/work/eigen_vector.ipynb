```python
# quiver.ipynb 

# doc:
# https://plotly.com/python/quiver-plots/
# https://plotly.com/python-api-reference/generated/plotly.figure_factory.create_quiver.html
import plotly.figure_factory as ff
import numpy as np

# 
# meshgrid cf: https://deepage.net/features/numpy-meshgrid.html
x, y = np.meshgrid(np.arange(-4, 4, .5), np.arange(-4, 4, .5))

# 
row_1, row_2 = np.array([[2.0, 1.0],
                         [1.0, 1.0]])

# 
# FIXME 
# u,v 
a, b = row_1
c, d = row_2
u= a*x + b*y
v= c*x + d*y 

# _u,_v 
_u = u-x
_v = v-y

# 
fig = ff.create_quiver(x, y, _u, _v,
                       scale=1, # 
                       arrow_scale=0.01 # (?)
                      )
fig.update_layout(
    margin=dict(l=20, r=20, t=20, b=20)
)
# fig.update_xaxes(showgrid=False)
fig.show()


# TODO: https://plotly.com/python/sliders/
```
